```
from google.colab import files
import os
import PyPDF2 
import textract
from cleantext import clean
import slate3k as slate

from pdfminer.pdfinterp import PDFResourceManager, process_pdf
from pdfminer.converter import TextConverter
from pdfminer.layout import LAParams
import io
# from nltk.tokenize import word_tokenize
# from nltk.corpus import stopwords

# open path get all school names
# go into each school and get pdfs
# convert pdf to text
# create new folder called "text"
```

```
def convert_pdf(pdfFile, retstr):
    password = ''
    pagenos = set()
    maxpages = 0
    laparams = LAParams()
    rsrcmgr = PDFResourceManager()
    device = TextConverter(rsrcmgr, retstr, laparams=laparams)    
    process_pdf(rsrcmgr, device, pdfFile, pagenos, maxpages=maxpages, password=password, check_extractable=True)
    device.close()
    return retstr
```
