```
import os
import pickle
import matplotlib.pylab as plt
```

```
#list of words we are looking for
list_of_words = ["machinelearning","machine learning", "artificialinteligence","artificial inteligence"]
text_path = "/content/drive/My Drive/US College Catalogs/MA/text/"
list_of_schools = []
for x in os.listdir(text_path):
  list_of_schools.append(x)
```

```
for school in list_of_schools:
  obj_to_save = {}
  count = 0
  print(school)
  for year in os.listdir(text_path + school):
    #print(year)
    for i in range(2009,2021):
      if year.find(str(i)) == 0:
        obj_to_save[i] = 0
        year_file = open(text_path + school + "/" + year, "r",encoding = "ISO-8859-1")
        text = year_file.read()
        for keyword in list_of_words:
          obj_to_save[i] = obj_to_save[i] + text.count(keyword)
        count = count + 1
        year_file.close()
        break
  file = open(text_path + school + "/toplot.obj" , 'wb')
  pickle.dump(obj_to_save, file)
  file.close()
```
