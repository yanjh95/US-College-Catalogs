```
import os
import pickle
import matplotlib.pylab as plt
```

```
# gets list of all schools
text_path = "/content/drive/My Drive/US College Catalogs/MA/text/"
# path to save plots
plot_path = "/content/drive/My Drive/US College Catalogs/MA/plots/"
list_of_schools = []
for x in os.listdir(text_path):
  list_of_schools.append(x)
```

```
for school in list_of_schools:
  obj_saved = []
  count = 0
  print(school)
  file = open(text_path + school + "/toplot.obj" , 'rb')
  obj_saved = pickle.load(file)
  file.close()

  plot_years = []
  plot_count = []
  for i in range(2009,2021):
    plot_years.append(i)
    if i in obj_saved:
      plot_count.append(obj_saved[i])
    else:
      plot_count.append(0)

  plt.bar(plot_years,plot_count,color='b')
  plt.savefig(plot_path + school + ".png")
```
